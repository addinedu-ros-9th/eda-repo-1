# DB 에 집어넣기

```python
import mysql.connector

remote = mysql.connector.connect (
    host = "jinhyuk2db.c94oygiiukcg.ap-northeast-2.rds.amazonaws.com",
    port = 3306,
    user = "root",
    password = "jinhyuk2dacibul",
    database = "project1"
)

cursor = remote.cursor(buffered=True)


sql= """INSERT INto weather_daily_B VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""

for i,row in df_merged.iterrows(): 

    weather_date = row['일시']
    temp_avg = row['평균기온(℃)']
    temp_max = row['최고기온(℃)']
    temp_min = row['최저기온(℃)']
    temp_diff = row['일교차']
    rain = row['강수량(mm)']
    humid_avg = row['평균습도(%rh)']
    humid_min = row['최저습도(%rh)']
    sunlight_direct = row['일조합(hr)']
    sunlight_ratio = row['일조율(%)']
    radiation_total = row['일사합(MJ/m2)']
    wind_avg = row['평균풍속(m/s)']
    wind_max = row['최대풍속(m/s)']
    temp_sense = row['체감온도(°C)']

    content = (weather_date, temp_avg, temp_max, temp_min, temp_diff,
               rain, humid_avg, humid_min, 
               sunlight_direct, sunlight_ratio, radiation_total, wind_avg, wind_max, temp_sense)

    cursor.execute(sql, content)

    print(content)

    

remote.commit()
remote.close()
```
